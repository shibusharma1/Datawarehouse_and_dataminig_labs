```python
#Create the empty list and append data in it.
li = []
bol = True

while bol:
    new_li = input('Enter the Fruits items for the list: ')
    li.append(new_li)
    user_input = input('Do you want to insert more data (yes/no)? ').lower()
    if user_input == 'no':
        bol = False

print("List:", li)
```

```python
#search the fruit in the list
search_item = input('Enter the Fruit you want to search:')
for fruit in li:
    if(fruit == search_item):
        a = 1
    else:
        a = 0
        
if(a == 1):
    print(f'The search item {fruit} exists in fruits list')
else:
    print(f"The search item {fruit} doesn't exists in fruits list")     
    
```

```python
#Exception handling.
li = []
bol = True

try:
    while bol:
        new_li = int(input('Enter item for the list: '))
        li.append(new_li)
        user_input = input('Do you want to insert more data (yes/no)? ').lower()
        if user_input == 'no':
            bol = False

    print("List:", li)


    search_item = int(input('Enter the num you want to search:'))
    for num in li:
        if(num == search_item):
            a = 1
        else:
            a = 0
            
    if(a == 1):
        print(f'The search item {num} exists in number list')
    else:
        print(f"The search item {num} doesn't exists in number list")  
        
except Exception as e:
    print(f'Invalid input type: {e}') 
```
