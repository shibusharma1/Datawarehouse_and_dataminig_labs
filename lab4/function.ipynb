```python
# calculate the factorial using recursion
def fact(num):
    if num == 0:
        return 1
    else:
        return num * fact(num - 1)
    
num = int(input('Enter the number:'))
result = fact(num)
print("The factorial of the number is:", result)
```

```python
# Function that returns multiple values: square and cube of a number
def square_cube(n):
    return n**2, n**3

num = int(input('Enter the number:'))
square, cube = square_cube(num)
print(f"Square of {num} is {square}")
print(f"Cube of {num} is {cube}")
```

```python
# WAP to check whether a number is even or odd using Function
def odd_even(num):
    if(num % 2 == 0):
        return "Even"
    else:
        return "Odd"
    
num = int(input('Enter the number:'))
print(odd_even(num))
```
