```python
# Creating a simple class
class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def show_info(self):
        print(f"Brand: {self.brand}, Color: {self.color}")

# objects
car1 = Car("Toyota", "Red")
car2 = Car("Honda", "Blue")

car1.show_info()
car2.show_info()
```

```python
# Encapusaltion
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance   #private variable

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

# Using encapsulation
acc = BankAccount("Shibu", 5000)
acc.deposit(2000)
print(acc.get_balance())

# print(acc.__balance)  #This will give an error (private variable)
```

```python
#Inheritance
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):   # Dog inherits from Animal
    def speak(self):
        print("Dog barks")

# Using inheritance
a = Animal()
d = Dog()

a.speak()
d.speak()
```

```python
#Polymorphism (Same Method, Different Behavior)
class Bird:
    def fly(self):
        print("Bird can fly")

class Penguin:
    def fly(self):
        print("Penguin cannot fly")

# Polymorphism in action
for animal in (Bird(), Penguin()):
    animal.fly()
```

```python
#Abstraction
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side

# Using abstraction
sq = Square(5)
print("Area of Square:", sq.area())
```
