```python
# 1. Create a dictionary
my_dict = {
    'name': 'Shibu',
    'age': 25,
    'city': 'Biratnagar'
}

print("Original Dictionary:", my_dict)

# 2. Access keys
print("Keys:", my_dict.keys())

# 3. Access values
print("Values:", my_dict.values())

# 4. Create a nested dictionary
nested_dict = {
    'student1': {
        'name': 'Shibu',
        'grade': 'A'
    },
    'student2': {
        'name': 'Sham',
        'grade': 'B'
    }
}

print("\nNested Dictionary:", nested_dict)

# 5. Access key and value from nested dictionary
print("Keys in nested dictionary:", nested_dict.keys())
print("Values in nested dictionary:", nested_dict.values())

# Access specific value
print("Student1's Name:", nested_dict['student1']['name'])
print("Student2's Grade:", nested_dict['student2']['grade'])
```
