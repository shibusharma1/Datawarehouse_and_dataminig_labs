```python
str1 = '   Shibu Sharma    '
# The strip() method in Python is used to remove leading and trailing characters from a string. By default, it removes whitespace characters such as spaces, tabs, and newlines. However, it can also be used to remove other specified character
print(str1.strip(' '))
```
