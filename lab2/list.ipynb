```python
list1 = ['apple','ball','cat']
print(list1[0])
print(list1[1])
print(list1[2])

list2 = ['apple','ball','cat','dog']
# slice(start, stop, step)
slice_list = slice(2,3,1)
print(list2[slice_list])
print(list2[slice(1,3,1)])


list3 = ['Aeroplane','Bus','Car']
print(list3)
list3.append('bike')
print(list3)
list3.insert(1,'train')
print(list3)
list3.remove('Car')
print(list3)
```
